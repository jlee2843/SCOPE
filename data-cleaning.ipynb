```python
import pandas as pd
import numpy as np
```

```python
cof_sum = pd.read_csv("data-processed/stabilized_lasso_Summary.csv")
cof_sum.sort_values(by = "model.count", ascending=False, inplace=True)
df = cof_sum
df
```

```python
gene_coords = pd.read_csv("data-processed/gene_coords")
gene_coords = gene_coords.rename(columns={"ensembl_gene_id":"probe"}).drop(["Unnamed: 0"], axis=1)
gene_coords
```

```python
df = pd.merge(cof_sum, gene_coords, on="probe", left_index=False, how = "outer")
df
```

```python
cols = df.columns.to_list()
cols
```

```python
second = cols[1:4]
first = cols[0:1]
last = cols[-1:]
cols2 = first + last + second
cols2
df = df[cols2]
sheet1 = df
sheet1
```

```python
sheet1 = sheet1.rename(columns={"probe":"Probe",
                               "external_gene_name":"Gene Name",
                               "model.count":"Model Count",
                               "min.val":"Minimum Value",
                               "max.val":"Maximum Value"})
sheet1
```

```python
sheet1.to_csv("GSE98394_sheet1.csv", index=False)
```

```python
pathways = pd.read_csv("data-processed/GSE98394_Pathways.csv")
pathways
col1 = pathways.columns.unique().to_list()
col1
```

```python
cols = pathways.columns.to_list()
first = cols[11:13]
last = cols[-2:]
second = cols[0:4]
third = cols[5:6]
enrichment = cols[6:7]
fourth = cols[7:11]
overlap = cols[4:5]
cols2 = first + enrichment + overlap + second + third + fourth + last
pathways = pathways[cols2]
sheet2 = pathways
pathways
sheet2.to_csv("GSE98394_sheet2.csv")
```

```python
pathways.sort_values(by="geneSet")
pathways1 = pathways.set_index(["geneSet"])
pathways1
```

```python
pathways1 = pathways1.reset_index()
pathways1
```

```python
pathways1["Ratio"] = pathways1["overlap"] / pathways1["size"]
pathways1
```

```python
colnames = pathways1["pathway_class"].str.split(";", expand=True)
colnames
```

```python
colnames.columns = ["Pathway Class", "Pathway Subclass"]
colnames
```

```python
pathways2 = pd.concat([pathways1, colnames[["Pathway Class", "Pathway Subclass"]]], axis=1)
pathways2
```

```python
pathways3 = pathways2[["geneSet", "description", "Pathway Class", "Pathway Subclass", "Core Gene Name", "size", "Ratio"]]
pathways3
```

```python
pathsize = pathways3.groupby(["geneSet", "description", "Pathway Class", "Pathway Subclass"])["size"].mean()
pathsize = pathsize.to_frame().reset_index()
pathsize
```

```python
pivoted = pathways3.pivot(index="geneSet", columns="Core Gene Name", values="Ratio")
pivoted = pivoted.reset_index()
pivoted = pivoted.rename_axis(None, axis=1)
pivoted
```

```python
sheet3 = pd.merge(pivoted, pathsize, on="geneSet")
sheet3
sheetname = sheet3.columns.to_list()
second = sheetname[-4:-1]
third = sheetname[-1:]
first = sheetname[0:1]
fourth = sheetname[1:5]
sheetname = first + second + third + fourth
sheetname
sheet3 = sheet3[sheetname]
sheet3 = sheet3.rename(columns = {"geneSet": "GeneSet",
                                 "size": "Pathway Size",
                                 "description": "Pathway Name"})
sheet3 = sheet3.rename(columns = {np.nan:"ENSG00000152022"})
cols = ["ENSG00000152022", "DISP1", "FGGY", "NADK"]
sheet3[cols] = sheet3[cols].fillna(0)
sheet3
```

```python
sheet3["POS"] = sheet3["ENSG00000152022"] + sheet3["DISP1"] + sheet3["FGGY"] + sheet3["NADK"]
sheet3
```

```python
sheet3.dtypes
```

```python
sheet3["Responsive Core Genes"] = sheet3[["ENSG00000152022", "DISP1", "FGGY", "NADK"]].gt(0).sum(axis=1)
sheet3 = sheet3.sort_values(by=["POS"], ascending=False)
sheet3
```

```python
sheet3.to_csv("GSE98394_sheet2.csv", index=False)
```
